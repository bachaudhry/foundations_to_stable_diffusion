Here, we are extracting the bits of code from the previous NB in the series which add noise to our inputs and shall proceed to call it `noisify()`

````python
def noisify(x0, ᾱ):
    device = x0.device
    n = len(x0)
    t = torch.randint(0, n_steps, (n,), dtype=torch.long)
    ε = torch.randn(x0.shape, device=device)
    ᾱ_t = ᾱ[t].reshape(-1, 1, 1, 1).to(device)
    xt = ᾱ_t.sqrt()*x0 + (1-ᾱ_t).sqrt()*ε
    return (xt, t.to(device)), ε
````

````python
@torch.no_grad()
def sample(model, sz, alpha, alpha_bar, sigma, n_steps):
    device = next(model.parameters()).device
    x_t = torch.randn(sz, device=device)
    preds = []
    for t in reversed(range(n_steps)): # Return a reversed iterator
        t_batch = torch.full((x_t.shape[0],), t, device=device, dtype=torch.long)
        z = (torch.randn(x_t.shape) if t>0 else torch.zeros(x_t.shape)).to(device)
        ᾱ_t1 = alpha_bar[t-1] if t>0 else torch.tensor(1)
        b̄_t = 1 - alpha_bar[t]
        b̄_t1 = 1 - ᾱ_t1
        x_0_hat = ((x_t - b̄_t.sqrt() * learn.model((x_t, t_batch))) / alpha_bar[t].sqrt()).clamp(-1, 1)
        x_t = x_0_hat * ᾱ_t1.sqrt() * (1-alpha[t])/b̄_t + x_t * alpha[t].sqrt()*b̄_t1/b̄_t + sigma[t]*z
        preds.append(x_t.cpu())
    return preds
````

````python
# The Callback has fewer lines of code since we extracted noisify as a separate function.
# The key difference here is that we are inheriting from the Callback class and not TrainCB.
# So, we won't be using the predict() function here.
class DDPMCB(Callback):
    order = DeviceCB.order+1
    def __init__(self, n_steps, beta_min, beta_max):
        super().__init__()
        fc.store_attr()
        self.beta = torch.linspace(self.beta_min, self.beta_max, self.n_steps)
        self.α = 1. - self.beta 
        self.ᾱ = torch.cumprod(self.α, dim=0)
        self.σ = self.beta.sqrt()
    
    def before_batch(self, learn): learn.batch = noisify(learn.batch[0], self.ᾱ)

    def sample(self, model, sz): return sample(model, sz, self.α, self.ᾱ, self.σ, self.n_steps)
````

The MiniAi framework allows us the flexibility to replace models, their components and interfaces, with our own versions. 

````python
# In the previous NB predict() was passing ((xt, t),ε) to the model which then return the sample.
# Here we can inherit from the UNet2DModel class to pass in the model, specifically the forward()
# function.
class UNet(UNet2DModel):
    # Calling the original forward() function, and passing it *x instead of x
    # and return sample
    def forward(self, x): return super().forward(*x).sample
````

We need to be able to use higher learning rates. But before we can do that, it is important to note that - at least at the time this NB was being written - the Diffusers library doesn't implement initializations. `init_ddpm()` will allow us to flexibly initalize elements of the ResNet blocks.

````python
def init_ddpm(model):
    for o in model.down_blocks:
        for p in o.resnets:
            # Take every second convolutional layer and zero it out.
            # This makes sure the non-id path of resnets do nothing at first so they don't
            # cause problems later on.
            p.conv2.weight.data.zero_()
            # Refer to orthogonal inits for the weights this time around.
            for p in fc.L(o.downsamplers): init.orthogonal_(p.conv.weight)

    for o in model.up_blocks:
        # Repeat the above for the up blocks as well.
        for p in o.resnets: p.conv2.weight.data.zero_()
    # Zero out the weights of the last layer.
    model.conv_out.weight.data.zero_()
````

### **Mixed Precision**

Mixed precision was intorduced in the paper titled ["Mixed Precision Training"](https://arxiv.org/abs/1710.03740).The abstract states:

> We introduce a technique to train deep neural networks using half precision floating point numbers. In our technique, weights, activations and gradients are stored in IEEE half-precision format. Half-precision floating numbers have limited numerical range compared to single-precision numbers. We propose two techniques to handle this loss of information.
>
> Firstly, we recommend maintaining a single-precision copy of the weights that accumulates the gradients after each optimizer step. This single-precision copy is rounded to half-precision format during training. Secondly, we propose scaling the loss appropriately to handle the loss of information with half-precision gradients. We demonstrate that this approach works for a wide variety of models including convolution neural networks, recurrent neural networks and generative adversarial networks.

In the paper, the authors claimed a 2x reduction in the memory consumption of neural networks making use of this technique.

[NVIDIA has excellent documentation on this topic](https://docs.nvidia.com/deeplearning/performance/mixed-precision-training/index.html) which provides good details on various implementation strategies and best practices.

````python
# Mixed precision works best with large batches, so we are increasing the bs by 4 times.
# The trade off is that with increased bs, the model has fewer batches to learn from.
# We can offset this by increasing the learning rate and the number of epochs.
bs = 512
````

````python
# Playing around with the idea of flexibility, we can put noisify inside the collation function.
# This effectively replaces before_batch() in the DDPMCB callback.
def collate_ddpm(b): return noisify(default_collate(b)[xl], alpha_bar)

# For details on how the collation function works, refer to NB 06_datasets.
# Here we will pass train and test data straight to the DataLoader init and not the from_dds() method.
def dl_ddpm(ds): return DataLoader(ds, batch_size=bs, collate_fn=collate_ddpm, num_workers=8)
````

````python
# Now we don't need a DDPM callback to handle data loading anymore.
dls = DataLoaders(dl_ddpm(tds['train']), dl_ddpm(tds['test']))
````

Taking advantage of Mixed Precision to speed up the model's training. We will replace default 32 bit fp values (single precision) with 16 bit fp values in a selective manner.

The callback for `MixedPrecision` is based on [PyTorch's documentation on the subject](https://pytorch.org/docs/stable/notes/amp_examples.html), specifically referring to the section:

```
with autocast(device_type='cuda', dtype=torch.float16):
    output = model(input)
    loss = loss_fn(output, target)
```

For this to work, the `Learner` in NB 09_Learner, was edited to include the `after_predict`, `after_loss`, `after_backward` and `after_step` callbacks. This allows subsequent callbacks to insert code at any point in the training loop.

From here on out, mixed precision can be added to any model built using the MiniAi framework.

````python
class MixedPrecision(TrainCB):
    order = DeviceCB.order+10
    def before_fit(self, learn): self.scaler = torch.cuda.amp.GradScaler()

    def before_batch(self, learn):
        # Based on PyTorch's documentation on Mixed Precision
        # autocast() gets passed as an attribute.
        self.autocast = torch.autocast("cuda", dtype=torch.float16)
        # Call context manager before the batch begins, effectively becoming a with statement.
        self.autocast.__enter__()

    def after_loss(self, learn): 
        # Exit the context manager after calculating the loss.
        self.autocast.__exit__(None, None, None)

    def backward(self, learn): 
        # Replace backward() with PyTorch's scaler.scaler.backward() for loss scaling.
        self.scaler.scale(learn.loss).backward()

    def step(self, learn):
        # Replace step() with scaler.step() and scaler.update()
        self.scaler.step(learn.opt)
        self.scaler.update()
````

````python
# Increasing lr and epochs to account for mixed precision.
lr = 1e-2
epochs = 10
tmax = epochs * len(dls.train)
sched = partial(lr_scheduler.OneCycleLR, max_lr=lr, total_steps=tmax)
cbs = [DeviceCB(), MixedPrecision(), ProgressCB(plot=True), MetricsCB(), BatchSchedCB(sched)]
model = UNet(in_channels=1, out_channels=1, block_out_channels=(16, 32, 64, 128), norm_num_groups=8)
init_ddpm(model)
# Using the standard Learner since mixed precision uses the TrainCB.
learn = Learner(model, dls, nn.MSELoss(), lr=lr, cbs=cbs, opt_func=opt_func)
````

## Switching to **HuggingFace Accelerate**

Thankfully, some FastAI alums working at HuggingFace created the [Accelerate](https://huggingface.co/docs/accelerate/index) project which contains all the functionality needed to implement techniques liked mixed precision training to speed up training times.

This library / project will allow us to use multi-GPU training for more intensive tasks.

For this section, we will decouple alot of the elements of DDPMCB.

````python
from accelerate import Accelerator
````

````python
class AccelerateCB(TrainCB): # Inherit from TrainCB since backward() will need to be changed
    order = DeviceCB.order+10
    def __init__(self, n_inp=1, mixed_precision="fp16"):
        super().__init__(n_inp=n_inp)
        self.acc = Accelerator(mixed_precision=mixed_precision)

    def before_fit(self, learn):
        # Wrap modules in the accelerator
        learn.model, learn.opt, learn.dls.train, learn.dls.valid = self.acc.prepare(
            learn.model, learn.opt, learn.dls.train, learn.dls.valid)

    def backward(self, learn): 
        # We will also replace the standard loss function i.e. loss.backward() with acc.backward(learn.loss)
        self.acc.backward(learn.loss)
````

````python
def noisify(x0, ᾱ):
    device = x0.device
    n = len(x0)
    t = torch.randint(0, n_steps, (n,), dtype=torch.long)
    ε = torch.randn(x0.shape, device=device)
    ᾱ_t = ᾱ[t].reshape(-1, 1, 1, 1).to(device)
    xt = ᾱ_t.sqrt()*x0 + (1-ᾱ_t).sqrt()*ε
    # Change to return a tuple instead of tuple of tuples, so we don't have to change anything about
    # the model, training callback etc.
    return xt, t.to(device), ε
````

````python
dls = DataLoaders(dl_ddpm(tds['train']), dl_ddpm(tds['test']))
````

````python
class DDPMCB2(Callback):
    # After predictions are complete, replace them with .sample.
    def after_predict(self, learn): learn.preds = learn.preds.sample
````

````python
model = UNet2DModel(in_channels=1, out_channels=1, block_out_channels=(16, 32, 64, 128),
                   norm_num_groups=8)
init_ddpm(model)
# Alterations to the TrainCB, which allowed more than one input can now be put to use since
# we are using accelerate.
cbs = [DDPMCB2(), DeviceCB(), ProgressCB(plot=True), MetricsCB(), BatchSchedCB(sched), AccelerateCB(n_inp=2)]
learn = Learner(model, dls, nn.MSELoss(), lr=lr, cbs=cbs, opt_func=opt_func)
````

````python
learn.fit(epochs)
````
